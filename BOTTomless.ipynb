```python
# resample
from pandas import read_csv
from pandas import datetime

series = read_csv('diagramm.csv', delimiter=";", decimal=",", header=0, parse_dates=[0], index_col=0, squeeze=True )
upsampled = series.resample('D')
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(32))
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = interpolated

x = a.index
y1 = a["Sonnenstunden"]
y2 = a[" Monatsgradtage"]

fig, ax = plt.subplots(figsize=(10,5),dpi=100)
ax.plot(x, y1,"black", linestyle="dashed", label = "Sonnenstunden")

ax.fill_between(x, y1, y2, where=y2>y1, facecolor='green',alpha=0.3)
ax.fill_between(x, y1, y2, where=y2<=y1, facecolor='red',alpha=0.3)
ax.set_ylabel("Sonnenstunden pro Monat",fontsize=12)


ax2 = ax.twinx()
ax2.plot(x, y2,"black", linestyle="-", label = "Monatsgradtage")
ax2.set_ylabel("Monatsgradtage",fontsize=12)

#ax.set_ylabel("Y")
#ax.set_title('Monatsgradtage - Sonnenstunden',fontsize=14)

ax.legend(loc=0)

fig.savefig("BOTTomless.png", dpi = 200)
```
